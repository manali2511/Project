```python
# Import libraries
import pandas as pd
import numpy as np

# (In Colab: Upload file "Data.csv")
from google.colab import files
uploaded = files.upload()

# Load dataset
df = pd.read_csv("Data.csv")

# Describe dataset
print("\nDataset Description:")
print(df.describe(include="all"))

# Shape of dataset
print("\nShape of dataset (rows, columns):", df.shape)

# Display first 3 rows
print("\nFirst 3 rows of dataset:")
print(df.head(3))
```

```python
# Check for missing values
print("\nMissing Values per Column:")
print(df.isnull().sum())

# Replace missing values in Age and Salary with column mean
# Handling Missing Values
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print("\nDataset after filling missing values:")
print(df)
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# OneHotEncoding Country
print("\nApplying OneHotEncoding to 'Country':")
df_encoded = pd.get_dummies(df, columns=['Country'])
print(df_encoded)

# LabelEncoding Purchased
label_encoder = LabelEncoder()
df_encoded['Purchased'] = label_encoder.fit_transform(df_encoded['Purchased'])

print("\nFinal Encoded Dataset:")
print(df_encoded)
```

```python
# 1. Import Libraries
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, Normalizer, Binarizer

# 2. Load Dataset
from google.colab import files
uploaded = files.upload()
```

```python


print("Original Data:")
print(df.head())

# -----------------------------------
# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

print("\nNumeric Columns:")
print(numeric_df.head())

# 1. Min-Max Scaling (0â€“1)
scaler = MinMaxScaler()
df_minmax = pd.DataFrame(scaler.fit_transform(numeric_df), columns=numeric_df.columns)

print("\nMin-Max Scaled Data (0-1):")
print(df_minmax.head())

# 2. Standardization (mean=0, std=1)
scaler = StandardScaler()
df_standard = pd.DataFrame(scaler.fit_transform(numeric_df), columns=numeric_df.columns)

print("\nStandardized Data (Z-score):")
print(df_standard.head())

# 3. Normalization (row-wise unit norm)
scaler = Normalizer()
df_normalized = pd.DataFrame(scaler.fit_transform(numeric_df), columns=numeric_df.columns)

print("\nNormalized Data (row-wise):")
print(df_normalized.head())

# 4. Binarization (threshold=0.5 for example)
scaler = Binarizer(threshold=0.5)
df_binarized = pd.DataFrame(scaler.fit_transform(numeric_df), columns=numeric_df.columns)

print("\nBinarized Data (threshold=0.5):")
print(df_binarized.head())
```

```python
df = pd.read_csv("Employee_Salary_Dataset.csv")
# -----------------------------
# Question 2: Display Data
# -----------------------------
print("First 5 rows of dataset:")
print(df.head())

# -----------------------------
# Question 3: Discretize Data
# -----------------------------
# Define bin edges and labels
bins = [0, 30000, 50000, 70000, 90000, 120000]
labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High']

# Create new column 'Salary_Bucket' using pd.cut()
df['Salary_Bucket'] = pd.cut(df['Salary'], bins=bins, labels=labels, include_lowest=True)

print("\nTop 10 rows after discretization:")
print(df.head(10))
```
