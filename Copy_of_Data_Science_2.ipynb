```python

# Load the dataset
df = pd.read_csv('SOCR-HeightWeight.csv')
```

```python
# Display first 10 rows
print("First 10 rows:\n", df.head(10))

# Display last 10 rows
print("\nLast 10 rows:\n", df.tail(10))

# Display 20 random rows
print("\nRandom 20 rows:\n", df.sample(20))
```

```python
print("\nShape of DataFrame:", df.shape)
print("Size of DataFrame:", df.size)
print("Data Types:\n", df.dtypes)
```

```python
print("\nStatistical Summary:\n", df.describe())
```

```python
print("\nTotal Rows (Observations):", df.shape[0])
print("Total Columns:", df.shape[1])
print("\nMissing Values Per Column:\n", df.isnull().sum())
```

```python
# Add BMI column
# Convert height and weight to metric units
df['Height(m)'] = df['Height(Inches)'] * 0.0254
df['Weight(kg)'] = df['Weight(Pounds)'] * 0.453592

# Calculate BMI
df['BMI'] = df['Weight(kg)'] / (df['Height(m)'] ** 2)

print("\nDataFrame with BMI:\n", df[['Height(m)', 'Weight(kg)', 'BMI']].head())
```

```python
# Maximum and Minimum BMI
print(f"\nMaximum BMI: {df['BMI'].max():.2f}")
print(f"Minimum BMI: {df['BMI'].min():.2f}")
```

```python
# Scatter plot - Height vs Weight
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(df['Height(m)'], df['Weight(kg)'], color='teal', alpha=0.6)
plt.title("Height vs Weight (in Metric Units)")
plt.xlabel("Height (meters)")
plt.ylabel("Weight (kilograms)")
plt.grid(True)
plt.show()
```
